````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import holidays
import warnings
warnings.filterwarnings('ignore')

from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
from prophet.plot import plot_cross_validation_metric
````

````python
from datetime import date

today     = '2026-02-27'
start     = '2010-01-01'

print(f"Downloading NIFTY data from {start} to {today}...")

df = yf.download("^NSEI", start=start, end=today)
df.columns = df.columns.get_level_values(0)
df.reset_index(inplace=True)

print(f"Downloaded {len(df)} rows")
print(f"Range: {df['Date'].min().date()} → {df['Date'].max().date()}")
print(f"Last close: {df['Close'].iloc[-1]:.2f}")
print(df.tail())
````

````python
prophet_df = df[['Date', 'Close']].copy()
prophet_df.columns = ['ds', 'y']
prophet_df['ds'] = pd.to_datetime(prophet_df['ds'])
prophet_df.dropna(inplace=True)

print(f"Prophet data ready!")
print(f"From: {prophet_df['ds'].iloc[0].date()}")
print(f"To:   {prophet_df['ds'].iloc[-1].date()}")
print(f"Last close: {prophet_df['y'].iloc[-1]:.2f}")
print(f"Total rows: {len(prophet_df)}")
````

````python
current_year = date.today().year

india_hols = holidays.India(years=range(2010, current_year + 2))

indian_holidays_df = pd.DataFrame({
    'holiday'      : 'indian_market_holiday',
    'ds'           : pd.to_datetime(list(india_hols.keys())),
    'lower_window' : -1,
    'upper_window' : 1,
}).sort_values('ds').reset_index(drop=True)

print(f"Total holidays loaded: {len(indian_holidays_df)}")
print(indian_holidays_df.head(10))
````

````python
model = Prophet(
    changepoint_prior_scale = 0.15,
    n_changepoints          = 30,
    yearly_seasonality      = True,
    weekly_seasonality      = True,
    daily_seasonality       = True,
    seasonality_prior_scale = 10,
    seasonality_mode        = 'multiplicative',
    holidays                = indian_holidays_df,
    holidays_prior_scale    = 10,
    interval_width          = 0.95,
)

print("⏳ Training model...")
model.fit(prophet_df)
print(f"Model trained on data up to: {prophet_df['ds'].iloc[-1].date()}")
print(f"Last known NIFTY close: {prophet_df['y'].iloc[-1]:.2f}")
````

````python
future   = model.make_future_dataframe(periods=365, freq='B')
forecast = model.predict(future)

print(f"Forecast generated!")
print(f"From: {forecast['ds'].min().date()}")
print(f"To:   {forecast['ds'].max().date()}")
````

````python
fig, ax = plt.subplots(figsize=(16, 6))

model.plot(forecast, ax=ax)

ax.set_title('NIFTY 50 — Prophet Forecast (365 Days Ahead)', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('NIFTY 50 Close Price')

last_date = prophet_df['ds'].max()
ax.axvline(x=last_date, color='red', linestyle='--', alpha=0.7)
ax.text(last_date, ax.get_ylim()[0] + 500,
        '← Actual  |  Forecast →',
        color='red', fontsize=10)

plt.tight_layout()
plt.show()
````

````python
metrics = performance_metrics(cv_results)

print("Model Performance:")
print(metrics[['horizon', 'mae', 'rmse', 'mape']].to_string())

fig3 = plot_cross_validation_metric(cv_results, metric='mape', figsize=(12, 5))
plt.title('MAPE vs Forecast Horizon')
plt.tight_layout()
plt.show()
````

Retune Prophet with Better Config

````python
prophet_df_log = prophet_df.copy()
prophet_df_log['y'] = np.log(prophet_df_log['y'])

print(f"Log transformed data ready")
print(f"Original last close: {prophet_df['y'].iloc[-1]:.2f}")
print(f"Log last close:      {prophet_df_log['y'].iloc[-1]:.4f}")

model_tuned = Prophet(
    changepoint_prior_scale = 0.20,
    changepoint_range = 0.95,
    n_changepoints = 40,
    yearly_seasonality = False,
    weekly_seasonality = False,
    daily_seasonality  = False,

    seasonality_prior_scale = 12,
    seasonality_mode = 'additive',
    holidays = indian_holidays_df,
    holidays_prior_scale = 10,
    interval_width = 0.95,
)
model_tuned.add_seasonality(
    name         = 'yearly',
    period       = 365.25,
    fourier_order = 15,
)

print("Training tuned Prophet on log data...")
model_tuned.fit(prophet_df_log)
print("Tuned Prophet trained!")
````

Get Prophet Predictions on Training Data


````python
future_tuned    = model_tuned.make_future_dataframe(
                    periods=365, freq='B')
forecast_tuned  = model_tuned.predict(future_tuned)
forecast_tuned['yhat']       = np.exp(forecast_tuned['yhat'])
forecast_tuned['yhat_lower'] = np.exp(forecast_tuned['yhat_lower'])
forecast_tuned['yhat_upper'] = np.exp(forecast_tuned['yhat_upper'])

print("Forecast generated and converted back from log scale!")
print(f"Forecast goes up to: {forecast_tuned['ds'].max().date()}")
merged = prophet_df.merge(
    forecast_tuned[['ds', 'yhat']],
    on='ds',
    how='inner'
)
merged['residual'] = merged['y'] - merged['yhat']

print(f"\nResidual Stats:")
print(f"Mean residual:  {merged['residual'].mean():.2f}")
print(f"Std residual:   {merged['residual'].std():.2f}")
print(f"Max residual:   {merged['residual'].max():.2f}")
print(f"Min residual:   {merged['residual'].min():.2f}")

print(f"\nSample (actual vs predicted vs residual):")
print(merged[['ds', 'y', 'yhat', 'residual']].tail(10).to_string(index=False))
````

Engineer Features for XGBoost

````python
features = merged.copy()
features = features.sort_values('ds').reset_index(drop=True)
features['day_of_week']  = features['ds'].dt.dayofweek
features['month']        = features['ds'].dt.month
features['quarter']      = features['ds'].dt.quarter
features['day_of_year']  = features['ds'].dt.dayofyear
features['week_of_year'] = features['ds'].dt.isocalendar().week.astype(int)
features['year']         = features['ds'].dt.year
features['lag_1']  = features['y'].shift(1)
features['lag_5']  = features['y'].shift(5)
features['lag_21'] = features['y'].shift(21)
features['lag_63'] = features['y'].shift(63)
features['rolling_mean_5']  = features['y'].rolling(5).mean()
features['rolling_mean_21'] = features['y'].rolling(21).mean()
features['rolling_std_5']   = features['y'].rolling(5).std()
features['rolling_std_21']  = features['y'].rolling(21).std()
features['momentum_5']  = features['y'] - features['lag_5']
features['momentum_21'] = features['y'] - features['lag_21']

features = features.dropna().reset_index(drop=True)

print(f"Features engineered!")
print(f"Total features: {len([c for c in features.columns if c not in ['ds', 'y', 'yhat', 'residual']])}")
print(f"Total rows after dropping NaN: {len(features)}")
print(f"\nFeature columns:")
feature_cols = ['day_of_week', 'month', 'quarter', 'day_of_year',
                'week_of_year', 'year', 'lag_1', 'lag_5', 'lag_21',
                'lag_63', 'rolling_mean_5', 'rolling_mean_21',
                'rolling_std_5', 'rolling_std_21',
                'momentum_5', 'momentum_21']
print(feature_cols)
````

Train XGBoost on Residuals

````python
from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings('ignore')
X = features[feature_cols]
y = features['residual']
split_idx   = int(len(X) * 0.80)
X_train     = X.iloc[:split_idx]
X_test      = X.iloc[split_idx:]
y_train     = y.iloc[:split_idx]
y_test      = y.iloc[split_idx:]

print(f"Train size: {len(X_train)} days")
print(f"Test size:  {len(X_test)} days")
print(f"Train up to: {features['ds'].iloc[split_idx-1].date()}")
print(f"Test from:   {features['ds'].iloc[split_idx].date()}")

xgb_model = XGBRegressor(
    n_estimators      = 500,
    learning_rate     = 0.05,
    max_depth         = 5,
    subsample         = 0.8,
    colsample_bytree  = 0.8,
    random_state      = 42,
    early_stopping_rounds = 50,
)

print("Training XGBoost on Prophet residuals...")
xgb_model.fit(
    X_train, y_train,
    eval_set        = [(X_test, y_test)],
    verbose         = False
)
print("XGBoost trained!")
import pandas as pd
importance = pd.DataFrame({
    'Feature'   : feature_cols,
    'Importance': xgb_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nWhich features matter most to XGBoost:")
print(importance.to_string(index=False))
````

 Combine Prophet + XGBoost Predictions



````python
test_data = features.iloc[split_idx:].copy()
prophet_pred = test_data['yhat'].values
xgb_correction = xgb_model.predict(X_test)
combined_pred = prophet_pred + xgb_correction
actual = test_data['y'].values

mae_prophet  = mean_absolute_error(actual, prophet_pred)
mae_combined = mean_absolute_error(actual, combined_pred)

mape_prophet  = np.mean(np.abs((actual - prophet_pred)  / actual)) * 100
mape_combined = np.mean(np.abs((actual - combined_pred) / actual)) * 100

improvement = ((mape_prophet - mape_combined) / mape_prophet) * 100

print("RESULTS — Prophet vs Prophet + XGBoost")
print("=" * 55)
print(f"{'Metric':<25} {'Prophet':>12} {'Combined':>12}")
print("-" * 55)
print(f"{'MAE (points)':<25} {mae_prophet:>12.2f} {mae_combined:>12.2f}")
print(f"{'MAPE (%)':<25} {mape_prophet:>12.2f} {mape_combined:>12.2f}")
print(f"{'Accuracy (%)':<25} {100-mape_prophet:>12.2f} {100-mape_combined:>12.2f}")
print("=" * 55)
print(f"\nImprovement: {improvement:.1f}% reduction in error")
print(f"Prophet alone:   {100-mape_prophet:.1f}% accurate")
print(f"Prophet+XGBoost: {100-mape_combined:.1f}% accurate")
````

````python
!pip install tensorflow -q

import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

print("All LSTM libraries imported!")
````

Prepare Data for LSTM

````python
SEQUENCE_LENGTH = 60
prices = prophet_df['y'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
prices_scaled = scaler.fit_transform(prices)

print(f"Data scaled to 0-1 range")
print(f"Original range: {prices.min():.2f} to {prices.max():.2f}")
print(f"Scaled range:   {prices_scaled.min():.4f} to {prices_scaled.max():.4f}")
print(f"Total data points: {len(prices_scaled)}")


def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

X_lstm, y_lstm = create_sequences(prices_scaled, SEQUENCE_LENGTH)
X_lstm = X_lstm.reshape(X_lstm.shape[0], X_lstm.shape[1], 1)

print(f"\nSequences created!")
print(f"X shape: {X_lstm.shape}")
print(f"   → {X_lstm.shape[0]} sequences")
print(f"   → each sequence = {X_lstm.shape[1]} days")
print(f"   → {X_lstm.shape[2]} feature per day (price)")
print(f"y shape: {y_lstm.shape}")
print(f"   → {y_lstm.shape[0]} targets (next day prices)")
````

Split Data & Build LSTM Model

````python
split_idx    = int(len(X_lstm) * 0.80)

X_train_lstm = X_lstm[:split_idx]
X_test_lstm  = X_lstm[split_idx:]
y_train_lstm = y_lstm[:split_idx]
y_test_lstm  = y_lstm[split_idx:]

print(f"Train size: {len(X_train_lstm)} sequences")
print(f"Test size:  {len(X_test_lstm)} sequences")

model_lstm = Sequential([

    LSTM(units       = 128,
         return_sequences = True,
         input_shape  = (SEQUENCE_LENGTH, 1)),
    Dropout(0.2),
    LSTM(units       = 64,
         return_sequences = False),
    Dropout(0.2),
    Dense(units = 32,
          activation = 'relu'),
    Dense(units = 1)
])

model_lstm.compile(
    optimizer = 'adam',
    loss      = 'mean_squared_error'
)

model_lstm.summary()
print("\nLSTM model built!")
````

Train LSTM

````python
early_stop = EarlyStopping(
    monitor   = 'val_loss',
    patience  = 15,
    restore_best_weights = True
)

print("Training LSTM (takes 3-5 mins)...")
print("Watch val_loss — should decrease over epochs")
print("If val_loss starts increasing = overfitting")
print()

history = model_lstm.fit(
    X_train_lstm, y_train_lstm,
    epochs          = 100,
    batch_size      = 32,
    validation_split = 0.10,
    callbacks       = [early_stop],
    verbose         = 1
)

print("\nLSTM training complete!")
print(f"Trained for {len(history.history['loss'])} epochs")
print(f"Final train loss:      {history.history['loss'][-1]:.6f}")
print(f"Final validation loss: {history.history['val_loss'][-1]:.6f}")
````

Compare All 3 Models

````python
test_dates = prophet_df['ds'].iloc[
    split_idx + SEQUENCE_LENGTH:
].reset_index(drop=True)

test_actual = prophet_df['y'].iloc[
    split_idx + SEQUENCE_LENGTH:
].values
lstm_pred_scaled = model_lstm.predict(X_test_lstm, verbose=0)
lstm_pred = scaler.inverse_transform(lstm_pred_scaled).flatten()
prophet_test = forecast_tuned[
    forecast_tuned['ds'].isin(test_dates)
]['yhat'].values
min_len      = min(len(test_actual), len(lstm_pred),
                   len(prophet_test))
test_actual  = test_actual[:min_len]
lstm_pred    = lstm_pred[:min_len]
prophet_test = prophet_test[:min_len]
features_test = features.iloc[split_idx:split_idx+min_len].copy()
if len(features_test) == min_len:
    xgb_correction_test = xgb_model.predict(
        features_test[feature_cols]
    )
    prophet_xgb_pred = prophet_test + xgb_correction_test
else:
    prophet_xgb_pred = prophet_test
w_prophet = 0.40
w_xgb     = 0.35
w_lstm    = 0.25

ensemble_pred = (
    w_prophet * prophet_test  +
    w_xgb     * prophet_xgb_pred +
    w_lstm    * lstm_pred
)
def mape(actual, predicted):
    return np.mean(np.abs((actual - predicted) / actual)) * 100

mape_prophet  = mape(test_actual, prophet_test)
mape_xgb      = mape(test_actual, prophet_xgb_pred)
mape_lstm     = mape(test_actual, lstm_pred)
mape_ensemble = mape(test_actual, ensemble_pred)

print("FINAL RESULTS — All Models Compared")
print("=" * 60)
print(f"{'Model':<30} {'MAPE':>8} {'Accuracy':>10}")
print("-" * 60)
print(f"{'Prophet tuned alone':<30} {mape_prophet:>7.2f}% {100-mape_prophet:>9.2f}%")
print(f"{'Prophet tuned + XGBoost':<30} {mape_xgb:>7.2f}% {100-mape_xgb:>9.2f}%")
print(f"{'LSTM alone':<30} {mape_lstm:>7.2f}% {100-mape_lstm:>9.2f}%")
print(f"{'Prophet + XGBoost + LSTM':<30} {mape_ensemble:>7.2f}% {100-mape_ensemble:>9.2f}%")
print("=" * 60)

best_mape  = min(mape_prophet, mape_xgb, mape_lstm, mape_ensemble)
best_model = ['Prophet tuned', 'Prophet+XGBoost',
              'LSTM', 'Full Ensemble'][
    [mape_prophet, mape_xgb, mape_lstm, mape_ensemble].index(best_mape)
]
print(f"\n🏆 Best model: {best_model} at {100-best_mape:.2f}% accuracy")
````

Previous days prediction

````python
specific_dates = pd.to_datetime(['2026-02-25', '2026-02-26', '2026-02-27'])
prophet_specific = forecast_tuned[
    forecast_tuned['ds'].isin(specific_dates)
][['ds', 'yhat']].copy().reset_index(drop=True)

lstm_preds = []
for target_date in specific_dates:
    idx = prophet_df[prophet_df['ds'] == target_date].index
    if len(idx) > 0:
        idx = idx[0]
        seq = prophet_df['y'].iloc[idx-60:idx].values.reshape(-1, 1)
        seq_scaled  = scaler.transform(seq)
        seq_input   = seq_scaled.reshape(1, 60, 1)
        pred_scaled = model_lstm.predict(seq_input, verbose=0)
        pred_price  = scaler.inverse_transform(pred_scaled)[0][0]
        lstm_preds.append(pred_price)
    else:
        lstm_preds.append(None)

xgb_corrections = []
for target_date in specific_dates:
    row_data = features[features['ds'] == target_date]
    if len(row_data) > 0:
        corr = xgb_model.predict(row_data[feature_cols])[0]
    else:
        corr = 0.0
    xgb_corrections.append(corr)

actual_df = yf.download("^NSEI",
                         start='2026-02-25',
                         end='2026-02-28')
actual_df.columns = actual_df.columns.get_level_values(0)
actual_df.reset_index(inplace=True)
actual_df['Date'] = pd.to_datetime(actual_df['Date'])
actual_df = actual_df[['Date', 'Close']].round(2)
actual_dict = dict(zip(actual_df['Date'], actual_df['Close']))

print("Feb 25, 26, 27 — All Models vs Actual")
print("=" * 95)
print(f"{'Day':<10} {'Date':<12} {'Prophet':>10} {'LSTM':>10} "
      f"{'Ensemble':>10} {'Actual':>10} {'Error%':>8} {'InBand':>8}")
print("-" * 95)

errors = []
for i, (_, row) in enumerate(prophet_specific.iterrows()):
    date_val    = row['ds']
    prophet_p   = row['yhat']
    lstm_p      = lstm_preds[i] if lstm_preds[i] else prophet_p
    xgb_corr    = xgb_corrections[i]
    prophet_xgb = prophet_p + xgb_corr

    final = (0.40 * prophet_p +
             0.35 * prophet_xgb +
             0.25 * lstm_p)

    actual    = actual_dict.get(date_val)
    bear_case = forecast_tuned[
        forecast_tuned['ds'] == date_val
    ]['yhat_lower'].values[0]
    bull_case = forecast_tuned[
        forecast_tuned['ds'] == date_val
    ]['yhat_upper'].values[0]

    if actual:
        error_pct = abs((actual - final) / actual) * 100
        in_band   = '✅' if bear_case <= actual <= bull_case else '❌'
        arrow     = "📈" if final > actual else "📉"
        errors.append(error_pct)

        print(f"{arrow} {date_val.day_name():<10} "
              f"{str(date_val.date()):<12} "
              f"{prophet_p:>10,.2f} "
              f"{lstm_p:>10,.2f} "
              f"{final:>10,.2f} "
              f"{actual:>10,.2f} "
              f"{error_pct:>7.2f}% "
              f"{in_band:>8}")
    else:
        print(f"⏳ {date_val.day_name():<10} "
              f"{str(date_val.date()):<12} "
              f"{prophet_p:>10,.2f} "
              f"{lstm_p:>10,.2f} "
              f"{final:>10,.2f} "
              f"{'N/A':>10} "
              f"{'N/A':>8} "
              f"{'⏳':>8}")

print("=" * 95)
if errors:
    avg_error = sum(errors) / len(errors)
    print(f"\nAverage Error:         {avg_error:.2f}%")
    print(f"Old basic model:       ~2.90%")
    print(f"Tuned Prophet:         ~1.65%")
    print(f"This ensemble:          {avg_error:.2f}%")
    print(f"Improvement vs old:    {((2.90-avg_error)/2.90*100):.1f}% better")
    print(f"Improvement vs tuned:  {((1.65-avg_error)/1.65*100):.1f}% better")
    print(f"\nBest model: {'Ensemble!' if avg_error < 1.65 else 'Tuned Prophet'}")
````

````python
import json
import numpy as np
import pandas as pd
import yfinance as yf
from prophet import Prophet
from datetime import date, timedelta
import warnings
warnings.filterwarnings('ignore')

today    = date.today()
tomorrow = today + timedelta(days=1)

print(f"Pulling latest NIFTY data up to {today}...")
df_fresh = yf.download("^NSEI",
                        start='2010-01-01',
                        end=str(tomorrow))
df_fresh.columns = df_fresh.columns.get_level_values(0)
df_fresh.reset_index(inplace=True)

prophet_df_latest = df_fresh[['Date', 'Close']].copy()
prophet_df_latest.columns = ['ds', 'y']
prophet_df_latest['ds'] = pd.to_datetime(prophet_df_latest['ds'])
prophet_df_latest.dropna(inplace=True)

print(f"Latest data point: {prophet_df_latest['ds'].iloc[-1].date()}")
print(f"Latest close:      {prophet_df_latest['y'].iloc[-1]:,.2f}")

print("\nRetraining tuned Prophet on latest data...")

prophet_df_log_latest = prophet_df_latest.copy()
prophet_df_log_latest['y'] = np.log(prophet_df_log_latest['y'])

model_fresh = Prophet(
    changepoint_prior_scale = 0.20,
    changepoint_range       = 0.95,
    n_changepoints          = 40,
    yearly_seasonality      = False,
    weekly_seasonality      = False,
    daily_seasonality       = False,
    seasonality_prior_scale = 12,
    seasonality_mode        = 'additive',
    holidays                = indian_holidays_df,
    holidays_prior_scale    = 10,
    interval_width          = 0.95,
)

model_fresh.add_seasonality(
    name          = 'yearly',
    period        = 365.25,
    fourier_order = 15,
)

model_fresh.fit(prophet_df_log_latest)
print(f"Model retrained on data up to: {prophet_df_latest['ds'].iloc[-1].date()}")

future_fresh    = model_fresh.make_future_dataframe(periods=30, freq='B')
forecast_fresh  = model_fresh.predict(future_fresh)
forecast_fresh['yhat']       = np.exp(forecast_fresh['yhat'])
forecast_fresh['yhat_lower'] = np.exp(forecast_fresh['yhat_lower'])
forecast_fresh['yhat_upper'] = np.exp(forecast_fresh['yhat_upper'])
print(f"Forecast generated up to: {forecast_fresh['ds'].max().date()}")

this_week = pd.bdate_range(
    start = str(today),
    periods = 5
)
target_dates = [str(d.date()) for d in this_week]
print(f"\nPredicting for: {target_dates}")
def fetch_recent_nifty_context_fresh():
    recent        = prophet_df_latest.tail(30).copy()
    last_price    = recent['y'].iloc[-1]
    price_30d_ago = recent['y'].iloc[0]
    price_7d_ago  = recent['y'].iloc[-5]
    change_30d    = ((last_price - price_30d_ago) / price_30d_ago) * 100
    change_7d     = ((last_price - price_7d_ago)  / price_7d_ago)  * 100
    volatility    = recent['y'].pct_change().std() * 100
    recent_returns = recent['y'].pct_change().tail(7).values * 100
    recent_returns = [f"{r:+.2f}%" for r in recent_returns if not np.isnan(r)]

    return f"""
    NIFTY 50 Latest Data:
    ─────────────────────────────
    Latest Close:         {last_price:,.2f}  ({prophet_df_latest['ds'].iloc[-1].date()})
    7 days ago:           {price_7d_ago:,.2f}
    30 days ago:          {price_30d_ago:,.2f}
    7-day change:         {change_7d:+.2f}%
    30-day change:        {change_30d:+.2f}%
    Recent volatility:    {volatility:.2f}% daily std
    Last 7 daily returns: {', '.join(recent_returns)}
    Trend:                {'Upward' if change_7d > 0 else 'Downward'}
    Market condition:     {'High volatility' if volatility > 1 else 'Low volatility'}
    """

def get_mistral_adjustment_fresh(prophet_prediction, target_date):
    print(f"Asking Mistral for {target_date}...")

    try:
        news_headlines = fetch_nifty_news()
        news_text = "\n".join([f"- {h}" for h in news_headlines])
    except:
        news_text = "- News unavailable"

    nifty_context = fetch_recent_nifty_context_fresh()

    prompt = f"""
You are an expert Indian stock market analyst specializing in NIFTY 50.

Analyze the context below and suggest a percentage adjustment
to the quantitative model's prediction.

═══════════════════════════════════════════════
MODEL PREDICTION
═══════════════════════════════════════════════
Target Date:       {target_date}
Prophet Predicted: {prophet_prediction:,.2f}

═══════════════════════════════════════════════
RECENT NIFTY DATA
═══════════════════════════════════════════════
{nifty_context}

═══════════════════════════════════════════════
LATEST NEWS HEADLINES
═══════════════════════════════════════════════
{news_text}

═══════════════════════════════════════════════
INSTRUCTIONS
═══════════════════════════════════════════════
- Be conservative (typical range -2% to +2%)
- Only adjust if there is CLEAR evidence
- 0% adjustment if no strong signal
- Consider: trend momentum, news sentiment,
  upcoming events, global cues

Respond in EXACTLY this JSON format only:
{{
  "adjustment_pct": 0.5,
  "direction": "up",
  "confidence": "medium",
  "reasoning": "One sentence explanation",
  "key_factor": "Single most important factor"
}}
"""

    response = client.chat.complete(
        model    = MISTRAL_MODEL,
        messages = [
            {
                "role"   : "system",
                "content": "You are a quantitative analyst. Respond with valid JSON only."
            },
            {
                "role"   : "user",
                "content": prompt
            }
        ],
        temperature = 0.1,
    )

    response_text = response.choices[0].message.content.strip()
    if "```json" in response_text:
        response_text = response_text.split("```json")[1].split("```")[0].strip()
    elif "```" in response_text:
        response_text = response_text.split("```")[1].split("```")[0].strip()

    return json.loads(response_text)

print("\nRunning Prophet + Mistral for this week...")
predictions = []

for target_date in target_dates:
    prophet_row = forecast_fresh[
        forecast_fresh['ds'] == pd.Timestamp(target_date)
    ]

    if len(prophet_row) == 0:
        print(f"No forecast for {target_date} — skipping")
        continue

    prophet_pred  = prophet_row['yhat'].values[0]
    prophet_lower = prophet_row['yhat_lower'].values[0]
    prophet_upper = prophet_row['yhat_upper'].values[0]

    try:
        mistral    = get_mistral_adjustment_fresh(prophet_pred, target_date)
        adj_pct    = mistral['adjustment_pct']
        direction  = mistral['direction']

        if direction == 'up':
            final_pred = prophet_pred * (1 + adj_pct/100)
        elif direction == 'down':
            final_pred = prophet_pred * (1 - adj_pct/100)
        else:
            final_pred = prophet_pred

        signed_adj = adj_pct if direction == 'up' else -adj_pct

    except Exception as e:
        print(f"Mistral failed for {target_date}: {e}")
        final_pred = prophet_pred
        signed_adj = 0.0
        mistral = {
            'confidence': 'N/A',
            'reasoning' : 'Mistral unavailable',
            'key_factor': 'N/A'
        }

    predictions.append({
        'date'        : pd.Timestamp(target_date),
        'prophet_pred': prophet_pred,
        'adj_pct'     : signed_adj,
        'final_pred'  : final_pred,
        'lower'       : prophet_lower,
        'upper'       : prophet_upper,
        'confidence'  : mistral['confidence'],
        'reasoning'   : mistral['reasoning'],
        'key_factor'  : mistral['key_factor']
    })
try:
    actual_df = yf.download("^NSEI",
                             start=str(today),
                             end=str(tomorrow))
    actual_df.columns = actual_df.columns.get_level_values(0)
    actual_df.reset_index(inplace=True)
    actual_df['Date'] = pd.to_datetime(actual_df['Date'])
    actual_dict = dict(zip(actual_df['Date'],
                           actual_df['Close'].round(2)))
except:
    actual_dict = {}

last_close      = prophet_df_latest['y'].iloc[-1]
errors_prophet  = []
errors_combined = []

print(f"\nProphet + Mistral — Week of {today}")
print(f"Base price (last known close): {last_close:,.2f}")
print("=" * 90)

for r in predictions:
    actual = actual_dict.get(r['date'])

    if actual:
        err_prophet  = abs((actual - r['prophet_pred']) / actual) * 100
        err_combined = abs((actual - r['final_pred'])   / actual) * 100
        in_band      = '✅' if r['lower'] <= actual <= r['upper'] else '❌'
        errors_prophet.append(err_prophet)
        errors_combined.append(err_combined)
        arrow        = "📈" if actual > last_close else "📉"

        print(f"\n{arrow} ACTUAL AVAILABLE")
        print(f"   {r['date'].day_name():<10} {r['date'].date()}")
        print(f"   Prophet predicted:  {r['prophet_pred']:>10,.2f}")
        print(f"   Mistral adjustment: {r['adj_pct']:>+9.2f}%  ({r['confidence']} confidence)")
        print(f"   Final prediction:   {r['final_pred']:>10,.2f}")
        print(f"   Actual:             {actual:>10,.2f}")
        print(f"   Prophet error:      {err_prophet:>9.2f}%")
        print(f"   Combined error:     {err_combined:>9.2f}%")
        print(f"   In band:            {in_band}")
        print(f"   Key factor:         {r['key_factor']}")
        print(f"   Reasoning:          {r['reasoning']}")

    else:
        change     = r['final_pred'] - last_close
        change_pct = (change / last_close) * 100
        arrow      = "📈" if change > 0 else "📉"

        print(f"\n{arrow} FUTURE PREDICTION")
        print(f"   {r['date'].day_name():<10} {r['date'].date()}")
        print(f"   Prophet predicted:  {r['prophet_pred']:>10,.2f}")
        print(f"   Mistral adjustment: {r['adj_pct']:>+9.2f}%  ({r['confidence']} confidence)")
        print(f"   Final prediction:   {r['final_pred']:>10,.2f}")
        print(f"   vs last close:      {change:>+10.2f} ({change_pct:>+.2f}%)")
        print(f"   Bear case:          {r['lower']:>10,.2f}")
        print(f"   Bull case:          {r['upper']:>10,.2f}")
        print(f"   Key factor:         {r['key_factor']}")
        print(f"   Reasoning:          {r['reasoning']}")

print("\n" + "=" * 90)
print(f"Week: {target_dates[0]} to {target_dates[-1]}")
print(f"Base price:  {last_close:,.2f}")
print(f"Days forecast: {len(predictions)}")

if errors_prophet and errors_combined:
    avg_err_prophet  = sum(errors_prophet)  / len(errors_prophet)
    avg_err_combined = sum(errors_combined) / len(errors_combined)
    improvement      = ((avg_err_prophet - avg_err_combined)
                        / avg_err_prophet * 100)
    print(f"\nAccuracy on verified days ({len(errors_prophet)}):")
    print(f"   Prophet alone error:      {avg_err_prophet:.2f}%")
    print(f"   Prophet + Mistral error:  {avg_err_combined:.2f}%")
    print(f"   Improvement:              {improvement:+.1f}%")
    verdict = "Mistral helped!" if avg_err_combined < avg_err_prophet else "Check manually"
    print(f"   Verdict:                  {verdict}")
else:
    print(f"\nAll future predictions — no actuals yet")
    print(f"   Run again tomorrow to verify accuracy!")
    print(f"\n   This week direction: "
          f"{'Bearish' if sum(r['adj_pct'] for r in predictions) < 0 else 'Bullish'}")
    print(f"   Avg Mistral adjustment: "
          f"{sum(r['adj_pct'] for r in predictions)/len(predictions):+.2f}%")
````
